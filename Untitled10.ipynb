```python
# Credit Card Fraud Detection - Imbalanced Classification
# Senior ML Engineer Implementation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from sklearn.metrics import (classification_report, confusion_matrix, roc_auc_score,
                           roc_curve, precision_recall_curve, f1_score, precision_score,
                           recall_score, auc)
from sklearn.decomposition import PCA
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTEENN
import warnings
warnings.filterwarnings('ignore')

# Set style for better plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

class FraudDetectionPipeline:
    """
    Complete pipeline for Credit Card Fraud Detection
    Handles imbalanced datasets with multiple techniques and models
    """

    def __init__(self, data_path=None):
        self.data_path = data_path
        self.df = None
        self.X_train = None
        self.X_test = None
        self.y_train = None
        self.y_test = None
        self.models = {}
        self.results = {}
        self.scaler = StandardScaler()

    def load_and_explore_data(self):
        """Load dataset and perform initial exploration"""
        print("=" * 60)
        print("CREDIT CARD FRAUD DETECTION ANALYSIS")
        print("=" * 60)

        # Load data
        if self.data_path:
            self.df = pd.read_csv(self.data_path)
        else:
            # For Colab users - upload file or use direct path
            try:
                from google.colab import files
                uploaded = files.upload()
                filename = list(uploaded.keys())[0]
                self.df = pd.read_csv(filename)
            except:
                # If not in Colab, try common filename
                self.df = pd.read_csv('creditcard.csv')

        print(f"\nðŸ“Š Dataset Shape: {self.df.shape}")
        print(f"ðŸ“‹ Dataset Info:")
        print(f"   - Total Transactions: {len(self.df):,}")
        print(f"   - Features: {self.df.shape[1] - 1}")
        print(f"   - Memory Usage: {self.df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

        # Check for missing values
        missing_values = self.df.isnull().sum().sum()
        print(f"   - Missing Values: {missing_values}")

        # Class distribution
        fraud_count = self.df['Class'].sum()
        total_count = len(self.df)
        fraud_percentage = (fraud_count / total_count) * 100

        print(f"\nðŸŽ¯ Class Distribution:")
        print(f"   - Normal Transactions: {total_count - fraud_count:,} ({100 - fraud_percentage:.4f}%)")
        print(f"   - Fraudulent Transactions: {fraud_count:,} ({fraud_percentage:.4f}%)")
        print(f"   - Imbalance Ratio: 1:{(total_count - fraud_count) // fraud_count}")

        return self.df

    def visualize_data_distribution(self):
        """Create comprehensive visualizations of the dataset"""
        plt.figure(figsize=(20, 15))

        # 1. Class Distribution
        plt.subplot(2, 3, 1)
        class_counts = self.df['Class'].value_counts()
        colors = ['#2ecc71', '#e74c3c']
        plt.pie(class_counts.values, labels=['Normal', 'Fraud'], autopct='%1.4f%%',
                colors=colors, startangle=90)
        plt.title('Transaction Class Distribution', fontsize=14, fontweight='bold')

        # 2. Transaction Amount Distribution
        plt.subplot(2, 3, 2)
        plt.hist(self.df[self.df['Class'] == 0]['Amount'], bins=50, alpha=0.7,
                label='Normal', color='#2ecc71', density=True)
        plt.hist(self.df[self.df['Class'] == 1]['Amount'], bins=50, alpha=0.7,
                label='Fraud', color='#e74c3c', density=True)
        plt.xlabel('Transaction Amount')
        plt.ylabel('Density')
        plt.title('Amount Distribution by Class', fontweight='bold')
        plt.legend()
        plt.yscale('log')

        # 3. Time Distribution
        plt.subplot(2, 3, 3)
        plt.hist(self.df[self.df['Class'] == 0]['Time'], bins=50, alpha=0.7,
                label='Normal', color='#2ecc71', density=True)
        plt.hist(self.df[self.df['Class'] == 1]['Time'], bins=50, alpha=0.7,
                label='Fraud', color='#e74c3c', density=True)
        plt.xlabel('Time')
        plt.ylabel('Density')
        plt.title('Time Distribution by Class', fontweight='bold')
        plt.legend()

        # 4. Correlation Matrix (key features with Class)
        plt.subplot(2, 3, 4)
        # Find features most correlated with Class
        corr_with_class = abs(self.df.corr()['Class']).sort_values(ascending=False)
        top_features = corr_with_class.head(8).index.tolist()  # Top 7 + Class
        corr_matrix = self.df[top_features].corr()
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', cbar_kws={'shrink': 0.8})
        plt.title('Correlation Matrix (Most Relevant Features)', fontweight='bold')

        # 5. Box plot for Amount by Class
        plt.subplot(2, 3, 5)
        sns.boxplot(data=self.df, x='Class', y='Amount')
        plt.title('Amount Distribution by Class', fontweight='bold')
        plt.yscale('log')

        # 6. Top features by correlation with fraud
        plt.subplot(2, 3, 6)
        # Calculate correlation with Class and get top features
        feature_corr = self.df.corr()['Class'].abs().sort_values(ascending=False)
        top_features = feature_corr.head(11)[1:11]  # Exclude 'Class' itself, take top 10

        colors = ['#e74c3c' if corr > 0.1 else '#3498db' for corr in top_features.values]
        plt.barh(range(len(top_features)), top_features.values, color=colors, alpha=0.7)
        plt.yticks(range(len(top_features)), top_features.index)
        plt.xlabel('Absolute Correlation with Fraud')
        plt.title('Top 10 Features Correlated with Fraud', fontweight='bold')
        plt.grid(axis='x', alpha=0.3)

        plt.tight_layout()
        plt.show()

    def prepare_data(self, test_size=0.3, random_state=42):
        """Prepare data for training"""
        print("\nðŸ”§ Preparing Data...")

        # Separate features and target
        X = self.df.drop(['Class'], axis=1)
        y = self.df['Class']

        # Split the data (IMPORTANT: Balance training data, NOT test data)
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
            X, y, test_size=test_size, random_state=random_state, stratify=y
        )

        # Scale the features
        self.X_train_scaled = self.scaler.fit_transform(self.X_train)
        self.X_test_scaled = self.scaler.transform(self.X_test)

        print(f"   - Training Set: {self.X_train.shape[0]:,} samples")
        print(f"   - Test Set: {self.X_test.shape[0]:,} samples")
        print(f"   - Training Fraud Rate: {self.y_train.mean():.4f}")
        print(f"   - Test Fraud Rate: {self.y_test.mean():.4f}")

        return self.X_train_scaled, self.X_test_scaled, self.y_train, self.y_test

    def apply_sampling_techniques(self):
        """Apply different sampling techniques to handle imbalanced data"""
        print("\nâš–ï¸ Applying Sampling Techniques...")

        sampling_techniques = {}

        # 1. SMOTE (Synthetic Minority Oversampling Technique)
        print("   - Applying SMOTE...")
        smote = SMOTE(random_state=42, k_neighbors=3)
        X_smote, y_smote = smote.fit_resample(self.X_train_scaled, self.y_train)
        sampling_techniques['SMOTE'] = (X_smote, y_smote)

        # 2. Random Under Sampling
        print("   - Applying Random Under Sampling...")
        rus = RandomUnderSampler(random_state=42)
        X_rus, y_rus = rus.fit_resample(self.X_train_scaled, self.y_train)
        sampling_techniques['RandomUnderSampler'] = (X_rus, y_rus)

        # 3. Combined SMOTE + ENN
        print("   - Applying SMOTE + ENN...")
        smote_enn = SMOTEENN(random_state=42)
        X_smote_enn, y_smote_enn = smote_enn.fit_resample(self.X_train_scaled, self.y_train)
        sampling_techniques['SMOTE_ENN'] = (X_smote_enn, y_smote_enn)

        # Print statistics
        for name, (X_resampled, y_resampled) in sampling_techniques.items():
            fraud_count = np.sum(y_resampled)
            total_count = len(y_resampled)
            print(f"     {name}: {total_count:,} samples ({fraud_count:,} fraud, {total_count-fraud_count:,} normal)")

        return sampling_techniques

    def train_models(self, sampling_techniques):
        """Train multiple models with different sampling techniques"""
        print("\nðŸš€ Training Models...")

        # Define models
        models_config = {
            'Logistic_Regression': LogisticRegression(random_state=42, max_iter=1000),
            'Random_Forest': RandomForestClassifier(random_state=42, n_estimators=100),
            'Isolation_Forest': IsolationForest(random_state=42, contamination=0.1)
        }

        # Train models with original data
        print("   - Training with Original Data...")
        for model_name, model in models_config.items():
            if model_name != 'Isolation_Forest':
                model.fit(self.X_train_scaled, self.y_train)
                self.models[f'{model_name}_Original'] = model
            else:
                # Isolation Forest is unsupervised
                model.fit(self.X_train_scaled)
                self.models[f'{model_name}_Original'] = model

        # Train models with different sampling techniques
        for sampling_name, (X_resampled, y_resampled) in sampling_techniques.items():
            print(f"   - Training with {sampling_name}...")
            for model_name, model_class in models_config.items():
                if model_name != 'Isolation_Forest':
                    model = model_class.__class__(**model_class.get_params())
                    model.fit(X_resampled, y_resampled)
                    self.models[f'{model_name}_{sampling_name}'] = model

    def evaluate_models(self):
        """Evaluate all trained models"""
        print("\nðŸ“Š Evaluating Models...")

        evaluation_results = []

        for model_name, model in self.models.items():
            print(f"   - Evaluating {model_name}...")

            if 'Isolation_Forest' in model_name:
                # Isolation Forest returns -1 for outliers, 1 for normal
                predictions = model.predict(self.X_test_scaled)
                y_pred = np.where(predictions == -1, 1, 0)  # Convert to 0/1
                y_pred_proba = model.decision_function(self.X_test_scaled)
                # Convert decision function to probability-like scores
                y_pred_proba = (y_pred_proba - y_pred_proba.min()) / (y_pred_proba.max() - y_pred_proba.min())
                y_pred_proba = 1 - y_pred_proba  # Flip so higher values indicate fraud
            else:
                y_pred = model.predict(self.X_test_scaled)
                y_pred_proba = model.predict_proba(self.X_test_scaled)[:, 1]

            # Calculate metrics
            precision = precision_score(self.y_test, y_pred)
            recall = recall_score(self.y_test, y_pred)
            f1 = f1_score(self.y_test, y_pred)
            roc_auc = roc_auc_score(self.y_test, y_pred_proba)

            # Precision-Recall AUC
            precision_curve, recall_curve, _ = precision_recall_curve(self.y_test, y_pred_proba)
            pr_auc = auc(recall_curve, precision_curve)

            evaluation_results.append({
                'Model': model_name,
                'Precision': precision,
                'Recall': recall,
                'F1-Score': f1,
                'ROC-AUC': roc_auc,
                'PR-AUC': pr_auc
            })

        # Create results DataFrame
        results_df = pd.DataFrame(evaluation_results)
        results_df = results_df.sort_values('F1-Score', ascending=False)

        print("\nðŸ† Model Performance Summary:")
        print(results_df.round(4).to_string(index=False))

        return results_df

    def visualize_model_performance(self, results_df):
        """Visualize model performance comparisons"""
        plt.figure(figsize=(20, 12))

        # 1. F1-Score Comparison
        plt.subplot(2, 3, 1)
        sns.barplot(data=results_df.head(10), y='Model', x='F1-Score', palette='viridis')
        plt.title('F1-Score Comparison (Top 10)', fontweight='bold')
        plt.xlabel('F1-Score')

        # 2. ROC-AUC vs PR-AUC
        plt.subplot(2, 3, 2)
        plt.scatter(results_df['ROC-AUC'], results_df['PR-AUC'],
                   c=results_df['F1-Score'], cmap='viridis', s=100, alpha=0.7)
        plt.xlabel('ROC-AUC')
        plt.ylabel('PR-AUC')
        plt.title('ROC-AUC vs PR-AUC', fontweight='bold')
        plt.colorbar(label='F1-Score')

        # 3. Precision vs Recall
        plt.subplot(2, 3, 3)
        plt.scatter(results_df['Recall'], results_df['Precision'],
                   c=results_df['F1-Score'], cmap='viridis', s=100, alpha=0.7)
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Precision vs Recall', fontweight='bold')
        plt.colorbar(label='F1-Score')

        # 4. ROC Curves for top 3 models
        plt.subplot(2, 3, 4)
        top_models = results_df.head(3)['Model'].tolist()

        for model_name in top_models:
            model = self.models[model_name]

            if 'Isolation_Forest' in model_name:
                predictions = model.decision_function(self.X_test_scaled)
                y_pred_proba = (predictions - predictions.min()) / (predictions.max() - predictions.min())
                y_pred_proba = 1 - y_pred_proba
            else:
                y_pred_proba = model.predict_proba(self.X_test_scaled)[:, 1]

            fpr, tpr, _ = roc_curve(self.y_test, y_pred_proba)
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.3f})')

        plt.plot([0, 1], [0, 1], 'k--', alpha=0.5)
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curves (Top 3 Models)', fontweight='bold')
        plt.legend()

        # 5. Precision-Recall Curves for top 3 models
        plt.subplot(2, 3, 5)
        for model_name in top_models:
            model = self.models[model_name]

            if 'Isolation_Forest' in model_name:
                predictions = model.decision_function(self.X_test_scaled)
                y_pred_proba = (predictions - predictions.min()) / (predictions.max() - predictions.min())
                y_pred_proba = 1 - y_pred_proba
            else:
                y_pred_proba = model.predict_proba(self.X_test_scaled)[:, 1]

            precision_curve, recall_curve, _ = precision_recall_curve(self.y_test, y_pred_proba)
            pr_auc = auc(recall_curve, precision_curve)
            plt.plot(recall_curve, precision_curve, label=f'{model_name} (AUC = {pr_auc:.3f})')

        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Precision-Recall Curves (Top 3 Models)', fontweight='bold')
        plt.legend()

        # 6. Confusion Matrix for best model
        plt.subplot(2, 3, 6)
        best_model_name = results_df.iloc[0]['Model']
        best_model = self.models[best_model_name]

        if 'Isolation_Forest' in best_model_name:
            predictions = best_model.predict(self.X_test_scaled)
            y_pred = np.where(predictions == -1, 1, 0)
        else:
            y_pred = best_model.predict(self.X_test_scaled)

        cm = confusion_matrix(self.y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=['Normal', 'Fraud'], yticklabels=['Normal', 'Fraud'])
        plt.title(f'Confusion Matrix - {best_model_name}', fontweight='bold')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')

        plt.tight_layout()
        plt.show()

        return best_model_name

    def pca_visualization(self):
        """Visualize fraud vs non-fraud transactions using PCA"""
        print("\nðŸ” Performing PCA Visualization...")

        plt.figure(figsize=(15, 10))

        # Apply PCA
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(self.X_train_scaled)

        # Create subplots
        plt.subplot(2, 2, 1)
        # Scatter plot of PCA components
        fraud_indices = self.y_train == 1
        normal_indices = self.y_train == 0

        plt.scatter(X_pca[normal_indices, 0], X_pca[normal_indices, 1],
                   c='#2ecc71', alpha=0.6, s=1, label='Normal')
        plt.scatter(X_pca[fraud_indices, 0], X_pca[fraud_indices, 1],
                   c='#e74c3c', alpha=0.8, s=10, label='Fraud')
        plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.3f} variance)')
        plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.3f} variance)')
        plt.title('PCA: Fraud vs Normal Transactions', fontweight='bold')
        plt.legend()

        # Explained variance
        plt.subplot(2, 2, 2)
        pca_full = PCA().fit(self.X_train_scaled)
        cumsum_variance = np.cumsum(pca_full.explained_variance_ratio_)
        plt.plot(range(1, len(cumsum_variance) + 1), cumsum_variance, 'bo-')
        plt.xlabel('Number of Components')
        plt.ylabel('Cumulative Explained Variance')
        plt.title('PCA Explained Variance', fontweight='bold')
        plt.axhline(y=0.95, color='r', linestyle='--', alpha=0.7, label='95% Variance')
        plt.legend()

        # 3D PCA with first 3 components
        ax = plt.subplot(2, 2, 3, projection='3d')
        pca_3d = PCA(n_components=3)
        X_pca_3d = pca_3d.fit_transform(self.X_train_scaled)

        ax.scatter(X_pca_3d[normal_indices, 0], X_pca_3d[normal_indices, 1],
                  X_pca_3d[normal_indices, 2], c='#2ecc71', alpha=0.1, s=1, label='Normal')
        ax.scatter(X_pca_3d[fraud_indices, 0], X_pca_3d[fraud_indices, 1],
                  X_pca_3d[fraud_indices, 2], c='#e74c3c', alpha=0.8, s=10, label='Fraud')
        ax.set_xlabel(f'PC1 ({pca_3d.explained_variance_ratio_[0]:.3f})')
        ax.set_ylabel(f'PC2 ({pca_3d.explained_variance_ratio_[1]:.3f})')
        ax.set_zlabel(f'PC3 ({pca_3d.explained_variance_ratio_[2]:.3f})')
        ax.set_title('3D PCA Visualization', fontweight='bold')
        ax.legend()

        # Feature importance for PCA components
        plt.subplot(2, 2, 4)
        # Create proper feature names for the creditcard dataset
        feature_names = ['Time'] + [f'V{i}' for i in range(1, 29)] + ['Amount']

        # Show top 10 features for PC1
        pc1_importance = np.abs(pca.components_[0])
        top_features_idx = np.argsort(pc1_importance)[-10:]
        top_features = [feature_names[i] for i in top_features_idx]
        top_importance = pc1_importance[top_features_idx]

        colors = ['#e74c3c' if 'V' in feat else '#f39c12' for feat in top_features]
        plt.barh(range(len(top_features)), top_importance, color=colors, alpha=0.8)
        plt.yticks(range(len(top_features)), top_features)
        plt.xlabel('Absolute Coefficient Value')
        plt.title('Top Features Contributing to PC1', fontweight='bold')
        plt.grid(axis='x', alpha=0.3)

        plt.tight_layout()
        plt.show()

        print(f"   - PC1 explains {pca.explained_variance_ratio_[0]:.3f} of variance")
        print(f"   - PC2 explains {pca.explained_variance_ratio_[1]:.3f} of variance")
        print(f"   - Total variance explained by PC1+PC2: {sum(pca.explained_variance_ratio_):.3f}")

    def detailed_analysis_best_model(self, best_model_name):
        """Provide detailed analysis of the best performing model"""
        print(f"\nðŸŽ¯ Detailed Analysis: {best_model_name}")
        print("=" * 60)

        best_model = self.models[best_model_name]

        if 'Isolation_Forest' in best_model_name:
            predictions = best_model.predict(self.X_test_scaled)
            y_pred = np.where(predictions == -1, 1, 0)
            y_pred_proba = best_model.decision_function(self.X_test_scaled)
            y_pred_proba = (y_pred_proba - y_pred_proba.min()) / (y_pred_proba.max() - y_pred_proba.min())
            y_pred_proba = 1 - y_pred_proba
        else:
            y_pred = best_model.predict(self.X_test_scaled)
            y_pred_proba = best_model.predict_proba(self.X_test_scaled)[:, 1]

        # Detailed classification report
        print("Classification Report:")
        print(classification_report(self.y_test, y_pred, target_names=['Normal', 'Fraud']))

        # Confusion Matrix Analysis
        cm = confusion_matrix(self.y_test, y_pred)
        tn, fp, fn, tp = cm.ravel()

        print(f"\nConfusion Matrix Analysis:")
        print(f"   - True Negatives (Correct Normal): {tn:,}")
        print(f"   - False Positives (Normal as Fraud): {fp:,}")
        print(f"   - False Negatives (Fraud as Normal): {fn:,}")
        print(f"   - True Positives (Correct Fraud): {tp:,}")

        # Business Impact Analysis
        print(f"\nBusiness Impact Analysis:")
        if fp > 0:
            false_positive_rate = fp / (fp + tn)
            print(f"   - False Positive Rate: {false_positive_rate:.4f} ({false_positive_rate*100:.2f}%)")
            print(f"     â†’ {fp:,} legitimate transactions flagged as fraud")

        if fn > 0:
            false_negative_rate = fn / (fn + tp)
            print(f"   - False Negative Rate: {false_negative_rate:.4f} ({false_negative_rate*100:.2f}%)")
            print(f"     â†’ {fn:,} fraudulent transactions missed")

        # Threshold Analysis
        print(f"\nThreshold Analysis:")
        thresholds = [0.1, 0.3, 0.5, 0.7, 0.9]
        print("Threshold | Precision | Recall | F1-Score")
        print("-" * 40)

        for threshold in thresholds:
            y_pred_thresh = (y_pred_proba >= threshold).astype(int)
            precision = precision_score(self.y_test, y_pred_thresh)
            recall = recall_score(self.y_test, y_pred_thresh)
            f1 = f1_score(self.y_test, y_pred_thresh)
            print(f"{threshold:^9} | {precision:^9.3f} | {recall:^6.3f} | {f1:^8.3f}")

    def run_complete_analysis(self):
        """Run the complete fraud detection analysis"""
        # Load and explore data
        self.load_and_explore_data()

        # Visualize data distribution
        self.visualize_data_distribution()

        # Prepare data
        self.prepare_data()

        # Apply sampling techniques
        sampling_techniques = self.apply_sampling_techniques()

        # Train models
        self.train_models(sampling_techniques)

        # Evaluate models
        results_df = self.evaluate_models()

        # Visualize performance
        best_model_name = self.visualize_model_performance(results_df)

        # PCA visualization
        self.pca_visualization()

        # Detailed analysis of best model
        self.detailed_analysis_best_model(best_model_name)

        print("\n" + "=" * 60)
        print("ANALYSIS COMPLETE!")
        print("=" * 60)
        print("\nKey Takeaways:")
        print("1. âœ… Successfully handled imbalanced dataset using SMOTE and RandomUnderSampler")
        print("2. âœ… Applied multiple models: Logistic Regression, Random Forest, Isolation Forest")
        print("3. âœ… Used appropriate metrics: Precision, Recall, F1-Score, ROC-AUC, PR-AUC")
        print("4. âœ… Visualized fraud patterns using PCA")
        print("5. âœ… Balanced training data while keeping test data intact")
        print(f"6. ðŸ† Best performing model: {best_model_name}")

        return results_df, best_model_name

# Example usage for Google Colab
if __name__ == "__main__":
    # Initialize the pipeline
    fraud_detector = FraudDetectionPipeline()

    # Run complete analysis
    # This will prompt for file upload in Colab
    results_df, best_model = fraud_detector.run_complete_analysis()

    # Optional: Save results
    # results_df.to_csv('fraud_detection_results.csv', index=False)

    print("\nðŸš€ Ready for production deployment!")
    print("ðŸ’¡ Consider hyperparameter tuning for even better performance.")
```
